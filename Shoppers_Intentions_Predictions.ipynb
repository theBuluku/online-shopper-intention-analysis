```r
#Loading the dataset for analysis
shoppers_data<- read.csv("http://bit.ly/EcommerceCustomersDataset")
```

```r
head(shoppers_data)
```

```r
#Checkinng the dimension of the dataset
dim(shoppers_data)
```

```r
#Checking for the names of the columns
attributes(shoppers_data)$names
```

```r
# Checking the datatypes for every column
columns = colnames(shoppers_data)
for (column in seq(length(colnames(shoppers_data)))){
    print(columns[column])
    print(class(shoppers_data[, column]))
    cat('\n')
}
```
